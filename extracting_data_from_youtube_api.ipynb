```python
#import libraries
import requests
import pandas as pd
import time
```

```python
#Keys
API_KEY = "***"
CHANNEL_ID = "UCW8Ews7tdKKkBT6GdtQaXvQ"
```

```python
def get_video_details(video_id):
    url_video_stats = "https://www.googleapis.com/youtube/v3/videos?id="+video_id+"&part=status&key="+API_KEY
    response_video_stats = requests.get(url_video_stats).json()
    #print(response_video_stats['items'][0]['status'].keys())
    uploadStatus = response_video_stats['items'][0]['status']['uploadStatus']
    privacyStatus = response_video_stats['items'][0]['status']['privacyStatus']
    license = response_video_stats['items'][0]['status']['license']
    madeForKids = response_video_stats['items'][0]['status']['madeForKids']
    return uploadStatus, privacyStatus, license, madeForKids
```

```python
def get_videos(df):
    pageToken = ""
    while 1:
      url = "https://www.googleapis.com/youtube/v3/search?key="+API_KEY+"&channelId="+CHANNEL_ID+"&part=snippet,id&order=date&maxResults=10000"+pageToken
      resp = requests.get(url).json()
      time.sleep(1) #give it a second before starting the for loop
      for v in resp["items"]:
        if v['id']['kind']=="youtube#video":
          vid_id = v['id']['videoId']
          title = str(v['snippet']['title']).replace("&amp;","")
          upload_date = str(v['snippet']['publishedAt']).split("T")[0]
          uploadStatus, privacyStatus, license, madeForKids = get_video_details(vid_id)
          df = df.append({'video_id':vid_id,'video_title':title,"upload_date":upload_date,"upload_status":uploadStatus,"privacy_status":privacyStatus,"license":license,"madeForKids":madeForKids},ignore_index=True)
          #print(vid_id,title,upload_date)
          #print(uploadStatus, privacyStatus, license, madeForKids,"\n")
      try:
        if response['nextPageToken'] != None: #if none, it means it reached the last page and break out of it
          pageToken = "pageToken=" + response['nextPageToken']
      except:
        break
    return df
```

```python
df2 = pd.DataFrame(columns=["video_id","video_title","upload_date","upload_status", "privacy_status", "license", "madeForKids"]) 
df2 = get_videos(df2)
```
